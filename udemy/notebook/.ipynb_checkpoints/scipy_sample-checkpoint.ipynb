```python
from scipy import misc, ndimage
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
raccoon = misc.face()
plt.imshow(raccoon)
type(raccoon)
```

```python
img = misc.ascent()
plt.imshow(img)
```

```python
gaussian_img = ndimage.gaussian_filter(input=raccoon, sigma = 10)
plt.imshow(gaussian_img)
```

```python
raccoon_img = Image.fromarray(raccoon)
img = np.array(raccoon_img.convert('L'),'f')
sigma = 5
img_x = ndimage.gaussian_filter(input=img, sigma=(sigma, sigma), order=(0,1))

plt.imshow(img_x, cmap='Greys_r')
```

```python
fig = plt.figure()

plt.scatter(np.random.rand(10), np.random.rand(10), s=100)
plt.axis('off')

fig.savefig('scatter.png')
```

```python
scatter_img = Image.open('scatter.png')
img = np.array(scatter_img.convert('L'))

img = 1 * (img < 128)
img
```

```python
im_open = ndimage.binary_opening(img)

plt.axis('off')
plt.imshow(im_open, cmap='Greys_r')
```

```python
label, num_features = ndimage.label(im_open)
num_features
```
